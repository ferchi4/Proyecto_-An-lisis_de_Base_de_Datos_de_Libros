## **2. Consultas SQL Realizadas**

## Consulta 1: Libros publicados después del 1 de enero de 2000

    Contamos con 819 libros publicados después del 1 de enero de 2000.

```python
Consulta_1= pd.read_sql("""
SELECT COUNT(book_id) AS libros_despues_2000 
FROM books 
WHERE publication_date > '2000-01-01';
""", 
con = engine)

print(Consulta_1)
```

## Consulta 2: Número de reseñas y calificación promedio por libro

    Podemos observar que el comportamiento de los usuarios varía significativamente entre los libros. Algunos libros tienen un alto número de reseñas y calificaciones promedio elevadas, lo que indica una fuerte interacción y satisfacción por parte de los lectores. Por otro lado, hay libros con pocas reseñas y calificaciones más bajas, lo que podría sugerir una menor popularidad o calidad percibida.

```python
Consulta_2 = pd.read_sql("""
SELECT 
    b.book_id,
    b.title,
    COUNT(DISTINCT rv.review_id) AS numero_resenas,
    AVG(rt.rating) AS calificacion_promedio
FROM books b
LEFT JOIN reviews rv ON b.book_id = rv.book_id
LEFT JOIN ratings rt ON b.book_id = rt.book_id
GROUP BY b.book_id, b.title
ORDER BY b.book_id;
""", con=engine)

print(Consulta_2)
```

## Consulta 3: Editorial con más libros de más de 50 páginas

    Con 42 libros de más de 50 páginas, la editorial "Penguin Books" es la que más contribuye a este segmento.

```python
Consulta_3 = pd.read_sql("""
SELECT 
    p.publisher,
    COUNT(b.book_id) AS numero_de_libros
FROM publishers AS p
JOIN books AS b ON p.publisher_id = b.publisher_id
WHERE b.num_pages > 50
GROUP BY p.publisher
ORDER BY numero_de_libros DESC
LIMIT 1;
""", con=engine)

print(Consulta_3)
```

## Consulta 4: Autor con mayor calificación promedio (libros con ≥50 calificaciones)

    J.K. Rowling es el autor con la calificación promedio más alta (4.28) entre aquellos cuyos libros han recibido al menos 50 calificaciones.

```python
Consulta_4 = pd.read_sql("""
SELECT 
    a.author,
    AVG(sub.avg_rating) AS rating_promedio_autor
FROM authors AS a
JOIN books AS b ON a.author_id = b.author_id
JOIN (
    -- Subconsulta para filtrar libros con >= 50 calificaciones
    SELECT 
        book_id, 
        AVG(rating) AS avg_rating 
    FROM ratings 
    GROUP BY book_id 
    HAVING COUNT(rating_id) >= 50
) AS sub ON b.book_id = sub.book_id
GROUP BY a.author
ORDER BY rating_promedio_autor DESC
LIMIT 1;
""", con=engine)

print(Consulta_4)
```

## Consulta 5: Promedio de reseñas textuales entre usuarios con >50 calificaciones

    24.33 es el promedio de reseñas textuales realizadas por usuarios que han dado más de 50 calificaciones.

```python
Consulta_5 = pd.read_sql("""
SELECT AVG(review_count) AS promedio_reseñas_texto
FROM (
    SELECT 
        username,
        COUNT(review_id) AS review_count
    FROM reviews
    WHERE username IN (
        -- Usuarios que calificaron más de 50 libros
        SELECT username
        FROM ratings
        GROUP BY username
        HAVING COUNT(rating_id) > 50
    )
    GROUP BY username
) AS sub;
""", con=engine)

print(Consulta_5)
```
